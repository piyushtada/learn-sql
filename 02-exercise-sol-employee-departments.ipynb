### 1. Retrieve the name and birth year of the employee’s child with code 350.

```python
%%sql

SELECT Name, BirthYear
FROM Dependents
WHERE Relationship = 'Child';
```

### 2. For each employee, retrieve the employee name and the name of the department where he works.


```python
%%sql

SELECT E.Name, D.Name
FROM Employees AS E, Departments AS D  
WHERE E.AssignedTo = D.Number;
```

### 3. Retrieve the names and birth years of female employees older than their supervisor.

```python
%%sql

SELECT u.Name, u.BirthYear
FROM DataSubordinates AS u
WHERE Sex = 'F'
    AND u.BirthYear > (
        SELECT s.BirthYear
        FROM DataSupervisors AS s
        WHERE u.SupervisedBy = s.SSN )
```

### 4. Retrieve the names of employees who do not have supervisors.

```python
%%sql
SELECT Name
FROM DataSubordinates
WHERE SupervisedBy IS NULL;
```

### 5. Retrieve the names of employees who work for the Research department (i.e. R&D) , and the location of the department.

```python
%%sql

SELECT e.Name, d.Location
FROM Employees AS e, Departments AS d
WHERE d.Name = 'R&D'
```

### 6. For every project located in Pisa, list the project number and name, the controlling department name, and the department manager’s name.

```python
%%sql

SELECT p.pNumber, p.pName, d.Name, e.Name
FROM Projects AS p, Departments AS d, Managers AS m, Employees AS e
WHERE p.ManagedBy = d.Number
AND m.Directs = d.Number
AND m.SSN = e.SSN
AND d.Location = 'Pisa'
```

### 7. Retrieve the names of employees who have no dependents.

```python
%%sql
SELECT e.Name
FROM employees As e, Dependents AS d
WHERE e.SSN = d.HouseHolder
```

### 8. Retrieve the names of supervisors who have at least one dependent.

```python
%%sql
SELECT e.Name
FROM Employees AS e, Supervisors AS s
WHERE e.SSN = s.SSN
        AND EXISTS (
        SELECT *
        FROM Dependents AS d
        WHERE d.HouseHolder = e.SSN)
```

```python
%%sql
SELECT Name
FROM DataSupervisors AS ds
WHERE EXISTS (
        SELECT *
        FROM Dependents AS d
        WHERE ds.SSN = d.HouseHolder)
```

### 9. For each employee retrieve the employees name and the name of the immediate supervisor.

```python
%%sql
SELECT ds.Name AS Employee, e.Name As Supervisor
FROM DataSubordinates AS ds, Employees AS e
WHERE ds.SupervisedBy = e.SSN;
```

### 10. Retrieve the names of employees who have a dependent with the same sex.

```python
%%sql
SELECT e.Name
FROM Employees AS e, Dependents AS d
WHERE e.SSN = d.HouseHolder
AND e.Sex = d.Sex;
```

### 11. For each employee retrieve the employees name and the social security number, the name of the project on which he works, and the name of the department that manages the project, sorted by the names of the department and the employee.

```python
%%sql

Select e.Name, e.SSN, p.PName AS Project_Name , d.Name AS Department_Name
FROM Employees AS e, Departments AS d, Projects As p, WorksOn AS w
WHERE e.AssignedTo = d.Number
AND d.Number = p.ManagedBy
AND e.SSN = w.employee
ORDER BY d.Name, e.Name;
```

### 12. For each project on which more than two employees work, retrieve the project number, the project name, and the number of employees who work on the project.

```python
%%sql
SELECT Project, COUNT(Employee)
FROM WorksOn AS w, Employees AS e
WHERE e.SSN = w.Employee
GROUP BY Project
HAVING COUNT(*) > 2;
```

### 13. For each project, retrieve the project number, the project name and the number of employees from department 10 (in my sample data we don't have department 10 so we look for department 3) who work on the project.

```python
%%sql

SELECT p.PNumber, p.PName, Count(*)
FROM Projects AS p, WorksOn AS w, Employees AS e
WHERE p.PNumber = w.Project
AND w.Employee = e.SSN
AND e.AssignedTo = 3
GROUP BY p.PNumber;
```
