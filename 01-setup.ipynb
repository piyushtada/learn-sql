# Creating materialized views

[PostgreSQL Materialized Views](https://www.postgresqltutorial.com/postgresql-materialized-views/)

We are creating views so that we can match our query with the results and have a simple query structure. 



```python
%%sql

CREATE MATERIALIZED VIEW DataSupervisors(
SSN, Name, BirthYear, Sex, AssignedTo, StartYear) AS
SELECT e.SSN, e.Name, e.BirthYear, e.Sex, 
    e.AssignedTo, s.StartYear
FROM Employees AS e, Supervisors AS s
WHERE e.SSN = s.SSN;
```

```python
%%sql

CREATE MATERIALIZED VIEW DataManagers (
    SSN, Name, BirthYear, Sex, AssignedTo, StartYear, Directs) AS
SELECT e.SSN, e.Name, e.BirthYear, e.Sex, e.AssignedTo,
    m.StartYear, m.Directs
FROM Employees AS e, Managers AS m
WHERE e.SSN = m.SSN;
```

```python
%%sql

CREATE MATERIALIZED VIEW DataSubordinates (
    SSN, Name, BirthYear, Sex, AssignedTo,SupervisedBy) AS
SELECT e.SSN, e.Name, e.BirthYear, e.Sex, e.AssignedTo, s.SupervisedBy
FROM Employees AS e, Subordinates AS s
WHERE e.SSN = s.SSN;
```
